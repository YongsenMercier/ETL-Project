```python
#import dependencies
import pandas as pd
import requests
import json
```

```python
pokemon = "pokemon.csv"
pokemon_df = pd.read_csv(pokemon, encoding="ISO-8859-1")
```

```python
pokemon_df
```

```python
column_list = [20,21,22,23,24,29,36,37,31]
pokemon_df=pokemon_df.drop(columns=pokemon_df.columns[column_list])
#pokemon_df.drop(columns=['base_egg_steps', 'base_happiness', 'base_total', ])
```

```python
pokemon_df2 = pokemon_df.set_index(["pokedex_number"])
pokemon_df2
```

```python
pokemon_df2 = pokemon_df2.iloc[0:386, :]
pokemon_df2
```

```python
pokemon_df3 = pokemon_df2[["name", "generation", "hp", "experience_growth", "abilities", "against_bug", "against_dark", "against_dragon", "against_electric",
                          "against_fairy", "against_fight", "against_fire", "against_flying", "against_ghost", "against_grass", "against_ground",
                           "against_ice", "against_normal", "against_poison", "against_psychic", "against_rock", "against_steel", "against_water", "attack",
                          "defense","height_m","sp_attack","sp_defense","speed","weight_kg","is_legendary"]]
pokemon_df3
```

```python
poke_name = pokemon_df3["name"]
```

```python
poke_name
```

```python
type_list1 = []
type_list2 = []

base = "https://pokeapi.co/api/v2/pokemon/"
try:
    for poke in poke_name:
        url = base + poke.lower() + "/"
        data = requests.get(url).json()
        type_list1.append(data["types"][0]["type"]['name'])
        print(f"{poke}")
        try:
            type_list2.append(data["types"][1]["type"]['name'])
        except:
            type_list2.append("N/A")
except:
    print("Something went wrong.")
```

```python
pokemon_df3["Type 2"] = type_list2
```
